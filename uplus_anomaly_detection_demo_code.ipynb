# 데이터 확인 및 간단한 전처리

```python
# 분포가 적당히 있는? 컬럼 가져오기 (값이 하나만 나오면 Drop)
max_val = round(len(results)*0.95)
filtered_col = []
for i in results.columns:
  # Max값의 Count가 95% 이상이면 Drop!
  if max(np.unique(results[i], return_counts = True)[1]) < max_val:
    filtered_col.append(i)
    print(i, end = ", ")
```

```python
# 필요한 컬럼만 남기기
filtered_df = results[filtered_col]
filtered_df = filtered_df.drop(columns = ['int64_field_0','Unnamed__0', 'index_date'])
```

```python
# 실습때는 X
# 조금 편하게 보기 위해서 필터 된 값으로 새로운 Table 만들어 저장!
import pandas_gbq # Pandas Dataframe을 빅쿼리 Table로 바로 저장
pandas_gbq.to_gbq(filtered_df, 'jc-gcp-project.LG_test.lg_uplus_poc_5_filtered', project_id='jc-gcp-project')
```

```python
## 저장한 테이블로 불러오기 (필요시 활용)
# %%bigquery filtered_df --project jc-gcp-project
# SELECT * FROM `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`
```

```python
# 주어진 data로 Test하기 위해 homecd_no 컬럼 그냥 제거
filtered_df = filtered_df.drop(['homecd_no'], axis = 1)

# Normal:0, ab:1로 대체
filtered_df.loc[filtered_df['label'] == 'normal', 'label'] = 0
filtered_df.loc[filtered_df['label'] == 'ab', 'label'] = 1
x_data = filtered_df.drop(['label'], axis = 1)
y_data = filtered_df.label

# Numeric value들은 Scaling 해주기
num_cols = ['ap_da01_cnt', 'ap_da02_cnt', 'ap_da10_cnt', 'ap_da12_cnt', 'ap_db06_cnt', 'ap_db07_cnt', 'ap_db08_cnt', 'ap_db09_cnt', 'ap_dc12_cnt', 'ap_dc13_cnt', 'ap_dc14_cnt', 'ap_da69_cnt', 'ap_da77_cnt', 'ap_da78_cnt', 'ap_da79_cnt', 'stb_hc_12_cnt', 'stb_hc_15_cnt', 'stb_hc_16_cnt', 'stb_hc_31_cnt']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_data[num_cols] = scaler.fit_transform(x_data[num_cols])

# Trainset, Testset 데이터 구분
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data)
```

```python
print('Y-값 Train set 비율 : ', np.unique(y_train, return_counts = True))
print('Y-값 Test set 비율 : ', np.unique(y_test, return_counts = True))
```

## XGBClassifier

```python
# xgbclassifier
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train._get_numeric_data(), y_train)
y_pred = model.predict(x_test._get_numeric_data())
```

## LSTM

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.activations import relu
from keras.optimizers import Adam
from keras import regularizers
from keras.callbacks import EarlyStopping
```

```python
# LSTM용 간단한 전처리
x_data = x_data.drop(['ctime'], axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data)
x_train = np.expand_dims(x_train, axis=1)
x_test = np.expand_dims(x_test, axis=1)
# y값 카테고리화
from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
def build_model():

  opt = Adam(learning_rate=0.01)

  model = Sequential()
  model.add(LSTM(12, dropout=0.7, recurrent_dropout=0.7))
  model.add(Dense(6, activation = 'relu'))
  model.add(Dropout(0.9))
  model.add(Dense(2, activation='softmax'))
  model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
  return model

model_lstm=build_model()
```

```python
es = EarlyStopping(monitor='val_loss', mode='min',patience=5)
history_lstm = model_lstm.fit(x_train, y_train, epochs=10,batch_size=3000,validation_data=(x_test,y_test),shuffle=False)
```

# BQML을 활용하여 위의 진행과정은 간단하게 수행해보기

### Logistic_reg: https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-glm

```python
## Logistic Regression
%%bigquery
CREATE OR REPLACE MODEL LG_test.logistic_regression
OPTIONS(
    model_type='LOGISTIC_REG',
    labels = ['label']) AS
SELECT * EXCEPT (ctime, homecd_no) FROM `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`;
```

```python
%%bigquery
SELECT
  *
FROM
  ML.PREDICT(MODEL `jc-gcp-project.LG_test.logistic_regression`, (
    SELECT *
    FROM
       `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`),
    STRUCT(0.1 AS threshold));
```

### Boosted Tree Classifier: https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-boosted-tree

```python
## Boosted Tree
%%bigquery
CREATE OR REPLACE MODEL LG_test.boosted_tree_classification_model
OPTIONS(
    model_type='BOOSTED_TREE_CLASSIFIER',
    labels = ['label']) AS
SELECT * EXCEPT (ctime, homecd_no) FROM `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`;
```

### DNN: https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-dnn-models

```python
## DNN Classifier
%%bigquery
CREATE OR REPLACE MODEL LG_test.dnn_classify_model
OPTIONS(
    model_type='DNN_CLASSIFIER',
    labels = ['label'],
    num_trials=5,
    max_parallel_trials=2,
    HPARAM_TUNING_ALGORITHM = 'VIZIER_DEFAULT'
    ) AS
    SELECT * EXCEPT (ctime, homecd_no) FROM `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`;
```

```python
%%bigquery
SELECT
  *
FROM
  ML.PREDICT(MODEL `jc-gcp-project.LG_test.dnn_classify_model`, (
    SELECT *
    FROM
       `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`),
    STRUCT(0.1 AS threshold));
```

### Wide-and-Deep: https://cloud.google.com/bigquery/docs/reference/standard-sql/bigqueryml-syntax-create-wnd-models

```python
%%bigquery
CREATE OR REPLACE MODEL LG_test.wide_and_deep_model
OPTIONS(MODEL_TYPE='DNN_LINEAR_COMBINED_CLASSIFIER',
        ACTIVATION_FN = 'RELU',
        BATCH_SIZE = 16,
        DROPOUT = 0.1,
        EARLY_STOP = FALSE,
        HIDDEN_UNITS = [128, 128, 128],
        INPUT_LABEL_COLS = ['label'],
        LEARN_RATE=0.001,
        MAX_ITERATIONS = 50,
        OPTIMIZER = 'ADAGRAD') AS
SELECT * EXCEPT (ctime, homecd_no) FROM `jc-gcp-project.LG_test.lg_uplus_poc_5_filtered`;
```
